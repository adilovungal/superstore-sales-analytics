```python
import pandas as pd

file_path = "../data/raw/Sample - Superstore.csv"

try:
    df = pd.read_csv(file_path, encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding="latin1")

print("Shape:", df.shape)

print("\nData types BEFORE conversion:\n", df.dtypes)

# Convert dates (important for warehouse)
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

print("\nData types AFTER conversion:\n", df.dtypes)

# Duplicate check
print("\nDuplicate rows:", df.duplicated().sum())

# Null check (key columns)
key_cols = ["Order ID", "Customer ID", "Product ID", "Sales", "Profit"]
print("\nNulls in key columns:\n", df[key_cols].isnull().sum())

# Basic anomaly check
print("\nNegative sales rows:", (df["Sales"] < 0).sum())
print("Negative profit rows:", (df["Profit"] < 0).sum())
```

```python
neg_profit_pct = (df["Profit"] < 0).mean() * 100
print(f"Negative profit percentage: {neg_profit_pct:.2f}%")
```
